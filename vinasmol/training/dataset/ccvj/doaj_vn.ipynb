# Permissively-licensed Vietnamese journals on DOAJ

This notebook finds Vietnamese journals from https://doaj.org and determines their [OAI-PMH](https://www.openarchives.org/pmh/) API URLs for paper harvesting.

````python
import pandas as pd
import requests
````

````python
in_vietnam = journals[journals["Country of publisher"] == "Viet Nam"]
````

## Filter by license

````python
compatible_licenses = ["CC BY", "CC BY-SA", "CC BY-NC", "CC BY-NC-SA"]
````

````python
usable = in_vietnam[(in_vietnam["Number of Article Records"] > 10) & in_vietnam["Journal license"].isin(compatible_licenses)]
usable_vietnamese = usable[usable["Languages in which the journal accepts manuscripts"].str.contains(".*Vietnamese.*")]
usable_english = usable[usable["Languages in which the journal accepts manuscripts"].str.contains(".*English.*")]
````

### Estimate number of articles by language

````python
num_records_maybe_vi = usable_vietnamese["Number of Article Records"].sum()
num_records_maybe_en = usable_english["Number of Article Records"].sum()
num_records = num_records_maybe_vi + num_records_maybe_en
print(
    f"Total number of records: {num_records}\n"
    f"Number of records possibly in Vietnamese: {num_records_maybe_vi}\n"
    f"Number of records possibly in English: {num_records_maybe_en}\n"
)
````

## Example request with DOAJ API

````python
from urllib.parse import quote
import json

def list_records_doaj_api(journal: dict) -> dict:
    """
    Make a request such as:

    ```bash
    curl -X 'GET' \\
      'https://doaj.org/api/search/articles/bibjson.journal.title%3A%22Journal%20of%20Advanced%20Engineering%20and%20Computation%22' \\
      -H 'accept: application/json'
    ```
    """
    title = journal["Journal title"]
    encoded_params = quote(f"bibjson.journal.title:\"{title}\"")
    url = f"https://doaj.org/api/search/articles/{encoded_params}"
    response = requests.get(
        url,
        headers=dict(accept="application/json")
    )
    response.raise_for_status()
    return json.loads(response.content)
````

## Infer the OAI APIs of each journal

````python
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def extract_meta_refresh_url(response: requests.Response) -> str | None:
    soup = BeautifulSoup(response.text, 'html.parser')
    meta_tag = soup.find('meta', attrs={'http-equiv': 'REFRESH'})
    if meta_tag:
        content = meta_tag['content']
        # Extract the refresh URL part
        _, url_part = content.split(';', 1)

        if url_part.strip().lower().startswith('url='):
            refresh_url = url_part.strip()[4:].strip()
            # Handle relative URLs
            if not refresh_url.startswith('http'):
                refresh_url = urljoin(response.url, refresh_url)

            return refresh_url
````

````python
import re

from util import auto_patch_ssl_verification

def infer_oai_api(journal_url: str, session: requests.Session) -> str | None:
    with auto_patch_ssl_verification(journal_url, session):
        # Resolve redirections
        response = session.get(journal_url)
    
    response.raise_for_status()

    refresh_url = extract_meta_refresh_url(response)
    base_url = refresh_url or response.url

    base_url = base_url.removesuffix("/index").removesuffix("/")

    if base_url.endswith(".vn"):
        # They might not host OAI-PMH themselves, but link to the VJOL OAI API
        vjol_oai_api = re.search(r"https://vjol.info.vn/index.php/[\w-]+", response.text)
        if vjol_oai_api is None:
            return
        return vjol_oai_api.group()

    # URL is [domain]/index.php/[journal] or [domain]/[journal]
    return f"{base_url}/oai"
    
````

````python
oai_apis: dict[str, str] = {}
````

````python
from tqdm.auto import tqdm

session = requests.Session()

for _, journal_title, journal_url in tqdm(
        usable[["Journal title", "Journal URL"]].itertuples(),
        total=len(usable),
    ):
    oai_apis[journal_url] = infer_oai_api(journal_url, session=session)
````

## Ensure that the inferred OAI-PMH endpoints work

````python
def probe_oai_api(oai_api_url: str):
    with auto_patch_ssl_verification(oai_api_url, session):
        # Resolve redirections
        response = session.get(urljoin(oai_api_url, "?verb=Identify"))
        response.raise_for_status()
````

````python
for oai_api_url in tqdm(oai_apis.values()):
    probe_oai_api(oai_api_url)
````

````python
from pathlib import Path

def save_oai_apis(oai_apis: dict[str, str], overwrite: bool = False):
    oai_apis_normalized = {
        (oai_api_url
            .removesuffix("/oai")
            .removeprefix("https://")
            .replace("/index.php/", "_"))
            .replace("/", "_"): oai_api_url
        for oai_api_url in oai_apis.values()
    }
    api_urls_json = Path("./data/ccvj/records/api_urls.json")

    if not api_urls_json.exists() or overwrite:
        api_urls_json.parent.mkdir(parents=True, exist_ok=True)
        api_urls_json.write_text(json.dumps(oai_apis_normalized, indent=2))
        print(f"Written {api_urls_json} to disk.")
    else:
        print(f"{api_urls_json} already exists.")
````

````python
save_oai_apis(oai_apis, overwrite=False)
````
